```python
##ATM Card Verification (Binary Search)
n = int(input())
cards = list(map(int, input().split()))
key = int(input())

low = 0
high = n - 1
found = False

while low <= high:
    mid = (low + high) // 2

    if cards[mid] == key:
        found = True
        break
    elif cards[mid] < key:
        low = mid + 1
    else:
        high = mid - 1

if found:
    print("Card Valid")
else:
    print("Invalid Card")
```
