```python
##Armstrong Numbers in a Range
start = int(input())
end = int(input())

for num in range(start, end + 1):
    power = len(str(num))
    total = 0
    temp = num

    while temp > 0:
        digit = temp % 10
        total += digit ** power
        temp //= 10

    if num == 0:
        total = 0

    if total == num:
        print(num, end=" ")
```

```python
##Automorphic Number
n = int(input())

square = n * n

if str(square).endswith(str(n)):
    print("Automorphic")
else:
    print("Not Automorphic")
```

```python
##Happy Number
n = int(input())

seen = set()

while n != 1 and n not in seen:
    seen.add(n)

    total = 0
    while n > 0:
        digit = n % 10
        total += digit * digit
        n //= 10

    n = total

if n == 1:
    print("Happy Number")
else:
    print("Not Happy Number")
```

```python
##Trapping Rain Water
n = int(input())
arr = list(map(int, input().split()))

left = [0] * n
right = [0] * n

left[0] = arr[0]
for i in range(1, n):
    left[i] = max(left[i - 1], arr[i])

right[n - 1] = arr[n - 1]
for i in range(n - 2, -1, -1):
    right[i] = max(right[i + 1], arr[i])

water = 0

for i in range(n):
    water += min(left[i], right[i]) - arr[i]

print(water)
```

```python
##Minimum number of platform
n = int(input())

arr = list(map(int, input().split()))
dep = list(map(int, input().split()))

arr.sort()
dep.sort()

i = j = 0
plat = ans = 0

while i < n and j < n:
    if arr[i] <= dep[j]:
        plat += 1
        ans = max(ans, plat)
        i += 1
    else:
        plat -= 1
        j += 1

print(ans)
```

```python
##Rotate Array by K Positions
n = int(input())

arr = list(map(int, input().split()))

k = int(input())

k = k % n

result = arr[-k:] + arr[:-k]

for i in result:
    print(i, end=" ")
```

```python
##Longest Common Prefix
n = int(input())

strings = []

for _ in range(n):
    strings.append(input())

prefix = strings[0]

for s in strings[1:]:

    while not s.startswith(prefix):

        prefix = prefix[:-1]

        if prefix == "":
            break

if prefix:
    print(prefix)
else:
    print(-1)
```

```python
##Minimum Window Substring
s = input()
t = input()

ans = ""

for i in range(len(s)):
    for j in range(i + 1, len(s) + 1):

        sub = s[i:j]

        ok = True
        for ch in t:
            if sub.count(ch) < t.count(ch):
                ok = False
                break

        if ok:
            if ans == "" or len(sub) < len(ans):
                ans = sub

print(ans)
```

```python
##Electricity Bill Calculation
units = int(input())

if units <= 100:
    bill = units * 2

elif units <= 300:
    bill = units * 3

else:
    bill = units * 5

print(bill)
```
