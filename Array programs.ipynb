```python
##Sum of Array Elements
n = int(input())
arr = list(map(int, input().split()))

print(sum(arr))
```

```python
## Maximum Element in Array
n = int(input())
arr = list(map(int, input().split()))

print(max(arr))
```

```python
##Minimum Element in Array
n = int(input())
arr = list(map(int, input().split()))

print(min(arr))
```

```python
##Reverse Array
n = int(input())
arr = list(map(int, input().split()))

for i in arr[::-1]:
    print(i, end=" ")
```

```python
##Search Element in Array
n = int(input())
arr = list(map(int, input().split()))
target = int(input())

if target in arr:
    print("Element Found")
else:
    print("Element Not Found")
```

```python
##Count Even and Odd Numbers
n = int(input())
arr = list(map(int, input().split()))

even = 0
odd = 0

for i in arr:
    if i % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even:", even)
print("Odd:", odd)
```

```python
##Sort Array in Ascending Order
n = int(input())
arr = list(map(int, input().split()))

arr.sort()

for i in arr:
    print(i, end=" ")
```

```python
##Second Largest Element
n = int(input())
arr = list(map(int, input().split()))

arr.sort()

print(arr[-2])
```

```python
##Remove Duplicate Elements
n = int(input())
arr = list(map(int, input().split()))

result = []

for i in arr:
    if i not in result:
        result.append(i)

for i in result:
    print(i, end=" ")
```
