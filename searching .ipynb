```python
n = int(input())

arr = []

for i in range(n):
    arr.append(int(input()))

x = int(input())

for i in range(n):
    if arr[i] == x:
        print("Found at Index", i)
        break
else:
    print("Product Not Found")
```

```python
n = int(input())

arr = []

for i in range(n):
    arr.append(int(input()))

found = False

for i in arr:
    if i % 2 == 0:
        found = True
        break

if found:
    print("Even Number Found")
else:
    print("Even Number Not Found")
```
