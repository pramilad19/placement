```python
##Half Pyramid Pattern
n = 5

for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end="")
    print()
```

```python
##Inverted Half Pyramid Pattern
n = 5

for i in range(n, 0, -1):
    for j in range(1, i + 1):
        print(j, end="")
    print()
```

```python
##Repeated Number Pyramid
n = 5

for i in range(1, n + 1):
    for j in range(i):
        print(i, end="")
    print()
```

```python
##Inverted Repeated Number Pyramid
n = 5

for i in range(n, 0, -1):
    for j in range(i):
        print(i, end="")
    print()
```

```python
##Floyd's Triangle
n = 5
num = 1

for i in range(1, n + 1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```

```python
##Reverse Number Triangle
n = 5

for i in range(1, n + 1):
    for j in range(i, 0, -1):
        print(j, end="")
    print()
```

```python
##Right Shifted Number Triangle
n = 5

for i in range(1, n + 1):
    for j in range(i, n + 1):
        print(j, end="")
    print()
```

```python
##Palindrome Number Pyramid
n = 5

for i in range(1, n + 1):

    print(" " * (n - i), end="")

    for j in range(1, i + 1):
        print(j, end="")

    for j in range(i - 1, 0, -1):
        print(j, end="")

    print()
```

```python
##Continuous Number Triangle
n = 4
num = 1

for i in range(1, n + 1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```

```python
##Binary Number Pattern
n = 5

for i in range(1, n + 1):
    for j in range(1, i + 1):
        print((i + j) % 2, end="")
    print()
```
