```python
#bubble sort
marks = [75, 90, 65, 85, 70]

n = len(marks)

for i in range(n):
    for j in range(n - i - 1):
        if marks[j] > marks[j + 1]:
            marks[j], marks[j + 1] = marks[j + 1], marks[j]

print(marks)
```

```python
#Product Price Sorting (Bubble Sort)
prices = [1200, 500, 2500, 800]

n = len(prices)

for i in range(n):
    for j in range(n - i - 1):
        if prices[j] > prices[j + 1]:
            prices[j], prices[j + 1] = prices[j + 1], prices[j]

print(prices)
```

```python
#Student Ranking (Selection Sort)
scores = [88, 72, 95, 60, 81]

n = len(scores)

for i in range(n):
    min_index = i

    for j in range(i + 1, n):
        if scores[j] < scores[min_index]:
            min_index = j

    scores[i], scores[min_index] = scores[min_index], scores[i]

print(scores)
```

```python
#Book ID Sorting (Selection Sort)
book_ids = [104, 101, 109, 102]

n = len(book_ids)

for i in range(n):
    min_index = i

    for j in range(i + 1, n):
        if book_ids[j] < book_ids[min_index]:
            min_index = j

    book_ids[i], book_ids[min_index] = book_ids[min_index], book_ids[i]

print(book_ids)
```

```python
ages = [35, 28, 42, 25, 31]

for i in range(len(ages)):
    min_pos = i

    for j in range(i + 1, len(ages)):
        if ages[j] < ages[min_pos]:
            min_pos = j

    ages[i], ages[min_pos] = ages[min_pos], ages[i]

print(ages)
```

```python
cards = [7, 3, 9, 1, 5]

for i in range(1, len(cards)):
    key = cards[i]
    j = i - 1

    while j >= 0 and cards[j] > key:
        cards[j + 1] = cards[j]
        j -= 1

    cards[j + 1] = key

print(cards)
```
